```python
import numpy as np 
import matplotlib.pyplot as plt 

K = float(input("Enter Strike Price (K): ")) 
P = float(input("Enter Premium (P): ")) 
S_min = float(input("Enter minimum Spot Price: ")) 
S_max = float(input("Enter maximum Spot Price: ")) 
Option=input("enter option type")
# Range of spot prices 
S = np.arange(S_min, S_max + 1) 

if P>=0:
    P=P
else:
    P=-P

buy_call = np.maximum(0, S - K) - P 
sell_call = P - np.maximum(0, S - K) 
buy_put = np.maximum(0, K - S) - P 
sell_put = P - np.maximum(0, K - S) 

plt.figure(figsize=(10, 6)) 
Option = Option.lower()
if Option=="call":
    plt.plot(S, buy_call, label="Buy Call", color="blue") 
    plt.plot(S, sell_call, label="Sell Call", color="red") 
elif Option=="put":
    plt.plot(S, buy_put, label="Buy Put", color="green") 
    plt.plot(S, sell_put, label="Sell Put", color="orange") 
plt.axhline(0, color="black", linewidth=0.8) 
# Profit/Loss line 
plt.axvline(K, color="gray", linestyle="--", label="Strike Price") 
plt.xlabel("Spot Price at Expiry") 
plt.ylabel("Profit / Loss") 
plt.title("Option Payoff Diagrams") 
plt.legend() 
plt.grid(True) 
plt.show() 
```
